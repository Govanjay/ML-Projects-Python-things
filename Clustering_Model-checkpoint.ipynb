# import libraries for clustering 

```python
import pandas as pd 
import matplotlib.pyplot as plt 
import numpy as np 
import seaborn as sns
from sklearn.datasets import load_iris 
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier 
```

# importing libraries for validating our model 

```python
from sklearn.metrics import confusion_matrix, classification_report 
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

# Load The Dataset 

```python
iris = load_iris()
```

```python
iris
```

# Now lets convert this data into row and columns using pandas 

```python
df = pd.DataFrame(iris.data, columns=iris.feature_names)
```

```python
df
```

# 




Now lets set our target

```python
df['target'] = iris.target
df['flower_name'] = df['target'].apply(lambda x: iris.target_names[x])
```

# define x and y or prepare the features

```python
x = df.drop(['target', 'flower_name'], axis=1 )  # features
y = df['target']  # the x is what we train but the y is what you predict
```

# lets scale our data

```python
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)
```

# Use the elbow method to determine the uptima number of cluster for k-means

```python
sse = []  # first create a list 
k_range = range(1,11)
for k in k_range:
    km = KMeans(n_clusters=k, random_state=42)
    km.fit(x_scaled)
    sse.append(km.inertia_)  # inertia means the distance 
```

# now lets plot

```python
plt.figure(figsize=(8, 5))
plt.plot(k_range, sse, marker='o')
plt.title('Elbow method for optima k')
plt.xlabel('number of clusters')
plt.ylabel('sun of the squared Distance (sse)')
plt.grid(True) # this add line to the chat
plt.show()
```

# let now apply the kmeans cliustering 

```python
kmeans = KMeans(n_clusters=3, random_state=42)
# them add the cluster colums to the dataset
df['cluster'] =kmeans.fit_predict(x_scaled)
```

```python
df
```

# lets visualised the clusters using PCA

```python
pca = PCA(n_components=2)
x_pca = pca.fit_transform(x_scaled)
df['pca1'] = x_pca[:,0]
df['pca2'] = x_pca[:,1]
df
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(data=df, x='pca1', y='pca2', hue='cluster', palette='viridis', s=100)
centroids_2d = pca.transform(kmeans.cluster_centers_)
plt.scatter(centroids_2d[:, 0], centroids_2d[:, 1], s=200, c='red', marker='X', label='centroids')
plt.legend()
plt.title('KMeans Cluster Visualised In 2D With Centroid')
plt.xlabel('pca1') # principal compoment analysis
plt.ylabel('pca2')
plt.grid(True)
plt.show()                
```

# Split data into training and testing 

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=1)
```

# train the model

```python
# Rememaber to always name the model the name of the model
knn = KNeighborsClassifier(n_neighbors=10)
knn.fit(x_train, y_train)
```

# Just check out the accuracy 0f the model

```python
accuracy = knn.score(x_test, y_test)
print(f'Model Accuracy: {accuracy:.2f}')
```

```python
# Now lets make a prediction and wherever you are making prediction please make use 
# of two dimensional arry [[]]. now lets use sample flower
sample = [[4.8, 3.0, 1.5, 0.3,]]
pred_class = knn.predict(sample)[0]
pred_name = iris.target_names[pred_class]
print(f'predicted class: {pred_name}')
```

# Now lets plot confusion matrix that show how many times our model fail

```python
y_pred = knn.predict(x_test)
cm = confusion_matrix(y_test, y_pred)
```

```python
plt.figure(figsize=(7,5))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.xlabel('The Predicted values')
plt.ylabel('The Actual value')
plt.title('confusion matrix')
plt.show()
```

# now lets print the classification report

```python
print('Classification Report: \n')
print(classification_report(y_test, y_pred, target_names=iris.target_names))
```
